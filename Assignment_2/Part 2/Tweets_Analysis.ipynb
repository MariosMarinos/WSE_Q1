```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# this method gives the descriptive statistics for one of the attributes
# based on the category variable relevance (depending 0 or 1.)
def getStatistics(df, attribute):
    attritube_relevant = df[df["relevanceJudge"]==1][attribute]
    attribute_non_relevant = df[df["relevanceJudge"]==0][attribute]
    print(attritube_relevant.describe())
    print(attribute_non_relevant.describe())
```

```python
# entities, entityTypes, tweetsPosted, sentiment, my feature : nFollowers
df = pd.read_csv('task2_data.csv')
# creating df for our 5 features. 
df_5_features = pd.DataFrame()
df_5_features['entities'] = df['#entities']
df_5_features['entityTypes'] = df['#entityTypes']
df_5_features['tweetsPosted'] = df['#tweetsPosted']
df_5_features['sentiment'] = df['sentiment']
df_5_features['hasURL'] = df['hasURL']
df_5_features['relevanceJudge'] = df['relevanceJudge']

print(df_5_features.head())
getStatistics(df_5_features, "hasURL")
```

```python
"""Scatter plot matrix to see the relations between
all different pairs of features."""
sns.set_theme(style="ticks")

sns.pairplot(df_5_features, hue="relevanceJudge")
```
